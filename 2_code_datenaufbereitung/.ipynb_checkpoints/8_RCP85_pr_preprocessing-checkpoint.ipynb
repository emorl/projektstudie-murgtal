# Datenaufbereitung RCP8.5 Niederschlag

###  1. Packete laden, Pfade definieren und Daten laden

```python
# Benötigte Packete
import xarray as xr
import geopandas as gpd
import rioxarray
import numpy as np
import pandas as pd
from pyproj import Geod
from shapely.geometry import Point
import os
from glob import glob

# Dateipfade definieren
nc_dir = "data/rcp85pr/"
#shp_path = "data/Murgtal_Outlines.shp"
shp_path = "Murgtal_boundingbox/Murgtal_boundingbox.shp"
output_csv = "data/murg_pr_rcp85_mean.csv"

# Laden der Shapefile für das Murgeinzugsgebiet, Shapefile in Koordinatensystem EPSG 4326 umwandeln
murg_shape = gpd.read_file(shp_path).to_crs("EPSG:4326")

#Liste erstellen für Tageswerte
all_data = []

# ALle Dateien holen sortiert nach Jahr
nc_files = sorted(glob(os.path.join(nc_dir, "*.nc")))
```

### 2. Loop zur Datenaufbereitung aller .nc Daten
Eine Shapefile kann nicht direkt auf eine .nc Datei angewendet werden. Shapefile = Vektoren, NC-Datei = Gitterzellen --> Maske schaut welche Gitterpunkte betroffen sind, sie hat das gleiche Format wie die Klimadaten in der .nc Datei

```python
# === 5. Dateien durchlaufen ===
for nc_path in nc_files:
    print("Verarbeite:", nc_path)
    
    # Datei öffnen
    pr_ds = xr.open_dataset(nc_path)
    pr = pr_ds["pr"]  # Niederschlagsvariable

    # 2D lat/lon-Arrays aus Klimamodelldaten nehmen
    lats = pr_ds["lat"].values
    lons = pr_ds["lon"].values

    # Maske erstellen für alle Gitterzellen, welche innerhalb des Murggebiets liegen
    mask = np.zeros_like(lats, dtype=bool) #Erstellt leere Maske
    for geom in murg_shape.geometry: #Liegt jeder Gitterzellenmittelpunkt in geom?
        mask |= np.vectorize(lambda x, y: geom.contains(Point(x, y)))(lons, lats) #Gitterzellen die in geom liegen werden auf TRUE gesetzt --> Maske erstellt
    mask_da = xr.DataArray(mask, dims=("rlat", "rlon")) # Maske als DataArray umwandeln, damit es mit den Klimamodelldaten übereinstimmt


    # Nur Werte des Klimamodells innerhalb des Murggebiets behalten
    pr_murg = pr.where(mask_da)

    mean_mmday = pr_murg.mean(dim=["rlat", "rlon"]) * 86400

    df = pd.DataFrame({
    "date": mean_mmday["time"].values,
    "precip_mm_day": mean_mmday.values
    })

    # An Gesamtliste hängen
    all_data.append(df)
```

### 3. Erstellen einer CSV mit allen Niederschlagsdaten von 2006 bis 2100

```python
# Alle Daten zusammenfügen
final_df = pd.concat(all_data, ignore_index=True)

# Als csv speichern
final_df.to_csv(output_csv, index=False)
print("Gesamte CSV gespeichert:", output_csv)
```

### 4. Tägliche Niederschlagsmenge visualisieren

```python

import pandas as pd
import matplotlib.pyplot as plt

# CSV laden
df = pd.read_csv("data/murg_pr_rcp85_mean_outlines.csv")

# Datum konvertieren (falls noch nicht)
df["date"] = pd.to_datetime(df["date"])

# Plot Niederschlag über Zeit
plt.figure(figsize=(12, 6))
plt.plot(df["date"], df["precip_mm_day"], color='b', label='Niederschlag (mm/Tag)')
plt.title("Täglicher Niederschlag im Murg-Einzugsgebiet (RCP85)")
plt.xlabel("Datum")
plt.ylabel("Niederschlag (mm/Tag)")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV laden
df = pd.read_csv("data/murg_pr_rcp85_fullperiod_weighted_mean.csv", parse_dates=["date"])
```

```python
df["year"] = df["date"].dt.year

# Jährliche Niederschlagssumme
annual_sum = df.groupby("year")["precip_mm_day"].sum()

# Top 10 niederschlagsreichste Jahre
print("Nasseste Jahre:")
print(annual_sum.sort_values(ascending=False).head(10))

# Plot
annual_sum.plot(kind="bar", figsize=(14,4), title="Jährlicher Niederschlag im Murgtal (mm/Jahr)")
plt.ylabel("Niederschlag [mm]")
plt.tight_layout()
plt.show()
```

```python
print("Trockenste Jahre:")
print(annual_sum.sort_values().head(10))
```

```python
df["month"] = df["date"].dt.month
monthly_avg = df.groupby("month")["precip_mm_day"].mean()

monthly_avg.plot(kind="bar", title="Mittlerer Niederschlag pro Monat (2006–2100)")
plt.ylabel("mm/Tag")
plt.xlabel("Monat")
plt.xticks(range(12), ["Jan", "Feb", "Mär", "Apr", "Mai", "Jun", "Jul", "Aug", "Sep", "Okt", "Nov", "Dez"])
plt.show()
```

```python
df["decade"] = (df["year"] // 10) * 10
decade_mean = df.groupby("decade")["precip_mm_day"].mean() * 365  # auf Jahreswerte umrechnen

decade_mean.plot(marker="o", title="Trend: Niederschlag pro Jahrzehnt")
plt.ylabel("Niederschlag [mm/Jahr]")
plt.grid(True)
plt.show()
```

```python
# Definitionen (du kannst anpassen)
dry_threshold = 1.0    # unter 1 mm/Tag = Trockentag
heavy_threshold = 20.0 # über 20 mm/Tag = Starkregen

# Jährliche Zählung
dry_days = df[df["precip_mm_day"] < dry_threshold].groupby(df["year"]).size()
heavy_days = df[df["precip_mm_day"] > heavy_threshold].groupby(df["year"]).size()

# Vergleich plotten
plt.plot(dry_days.index, dry_days.values, label="Trockentage (<1mm)")
plt.plot(heavy_days.index, heavy_days.values, label="Starkregentage (>20mm)")
plt.legend()
plt.title("Trockentage vs. Starkregentage pro Jahr")
plt.ylabel("Anzahl Tage")
plt.grid(True)
plt.show()
```
