# Datenaufbereitung RCP2.6 Temperatur

###  Packete laden, Pfade definieren und Daten laden

```python
# Benötigte Packete
import xarray as xr
import geopandas as gpd
import rioxarray
import numpy as np
import pandas as pd
from pyproj import Geod
from shapely.geometry import Point
import os
from glob import glob

# Dateipfade definieren
nc_dir = "data/rcp26temp/"
shp_path = "Murgtal_boundingbox/Murgtal_boundingbox.shp"
output_csv = "data/murg_tas_rcp26_mean.csv"

# Laden der Shapefile für das Murgeinzugsgebiet, Shapefile in Koordinatensystem EPSG 4326 umwandeln
murg_shape = gpd.read_file(shp_path).to_crs("EPSG:4326")

#Liste erstellen für Tageswerte
all_data = []

# ALle Dateien holen sortiert nach Jahr
nc_files = sorted(glob(os.path.join(nc_dir, "*.nc")))
```

### Loop zur Datenaufbereitung aller .nc Daten
Eine Shapefile kann nicht direkt auf eine .nc Datei angewendet werden. Shapefile = Vektoren, NC-Datei = Gitterzellen --> Maske schaut welche Gitterpunkte betroffen sind, sie hat das gleiche Format wie die Klimadaten in der .nc Datei

```python
for nc_path in nc_files:
    print("Verarbeite:", nc_path)
    
    # Datei öffnen
    tas_ds = xr.open_dataset(nc_path)
    tas = tas_ds["tas"]

    # 2D lat/lon extrahieren
    lats = tas_ds["lat"].values
    lons = tas_ds["lon"].values

    # Maske erstellen (Gitterzellen im Murggebiet)
    mask = np.zeros_like(lats, dtype=bool)
    for geom in murg_shape.geometry:
        mask |= np.vectorize(lambda x, y: geom.contains(Point(x, y)))(lons, lats)
    mask_da = xr.DataArray(mask, dims=("rlat", "rlon"))

    # Auf Murg beschränken
    tas_murg = tas.where(mask_da)

    # Mittelwert über Raum + Umrechnung in °C
    mean_temp_celsius = (tas_murg.mean(dim=["rlat", "rlon"]) - 273.15).rename("temp_day")

    # In DataFrame umwandeln
    df = mean_temp_celsius.to_dataframe().reset_index()
    df = df[["time", "temp_day"]].rename(columns={"time": "date"})

    all_data.append(df)

# === Alle Daten zusammenführen
final_df = pd.concat(all_data, ignore_index=True)

# === Bereinigen: Falls gemischte Datentypen → in float umwandeln
final_df["temp_day"] = pd.to_numeric(final_df["temp_day"], errors="coerce")

# === CSV speichern
final_df.to_csv(output_csv, index=False)
print("CSV gespeichert:", output_csv)
```

### Erstellen einer CSV mit allen Niederschlagsdaten von 2006 bis 2100

```python
final_df = pd.concat(all_data, ignore_index=True)
final_df.to_csv(output_csv, index=False)
print("Erste Zeilen aus CSV:")
print(pd.read_csv(output_csv).head())


```

### 4. Tägliche Niederschlagsmenge visualisieren

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV laden
df = pd.read_csv("data/murg_tas_rcp85_mean.csv")

# Datum konvertieren (falls noch nicht)
df["date"] = pd.to_datetime(df["date"])

# Plot Niederschlag über Zeit
plt.figure(figsize=(12, 6))
plt.plot(df["date"], df["temperature_day"], color='b', label='Grad Celsius pro Tag')
plt.title("Täglicher Niederschlag im Murg-Einzugsgebiet (RCP85)")
plt.xlabel("Datum")
plt.ylabel("Grad Celsius pro Tag")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Jahresmittel berechnen
df["year"] = df["date"].dt.year
annual_means = df.groupby("year")["temperature_day"].mean()

# Plot
plt.figure(figsize=(12, 5))
annual_means.plot(marker="o", color="darkred")
plt.title("Jährliche Mitteltemperatur im Murgtal (RCP8.5)")
plt.xlabel("Jahr")
plt.ylabel("Temperatur (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
rolling_mean = annual_means.rolling(window=10, center=True).mean()

plt.figure(figsize=(12, 5))
plt.plot(annual_means.index, annual_means, label="Jahresmittel", alpha=0.5)
plt.plot(rolling_mean.index, rolling_mean, label="10-Jahres-Gleitmittel", color="black")
plt.title("Langfristiger Temperaturtrend (RCP8.5)")
plt.xlabel("Jahr")
plt.ylabel("Temperatur (°C)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# CSV einlesen (falls noch nicht)
df = pd.read_csv("data/murg_tas_rcp85_mean.csv", parse_dates=["date"])

# Jahresmittel berechnen
df["year"] = df["date"].dt.year
annual_mean = df.groupby("year")["temperature_day"].mean()

# Lineare Regression (x = Jahr, y = Jahresmittel)
slope, intercept, r_value, p_value, std_err = linregress(annual_mean.index, annual_mean.values)

# Trendlinie berechnen
trendline = intercept + slope * annual_mean.index

# Plot
plt.figure(figsize=(12, 6))
plt.plot(annual_mean.index, annual_mean.values, label="Jahresmitteltemperatur", color="steelblue")
plt.plot(annual_mean.index, trendline, label=f"Trend: {slope*10:.2f} °C/Dekade", color="red", linestyle="--")
plt.xlabel("Jahr")
plt.ylabel("Temperatur (°C)")
plt.title("Klimatrend im Murgtal (RCP8.5)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Optional: Statistiken ausgeben
print("📈 Lineare Regressionsergebnisse:")
print(f"Anstieg pro Jahr: {slope:.4f} °C")
print(f"Anstieg pro Dekade: {slope*10:.4f} °C")
print(f"R²: {r_value**2:.3f}")
print(f"p-Wert: {p_value:.3e}")
```
