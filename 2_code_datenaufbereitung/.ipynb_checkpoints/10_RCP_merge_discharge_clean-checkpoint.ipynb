## CSV-Dateien der RCPs zusammenfügen

```python
import pandas as pd

# Einlesen mit Umbenennung der Spalten
rcp26_tas = pd.read_csv("data/murg_pr_rcp26_mean.csv").rename(columns={"precip_mm_day": "pr_day_rcp26"})
rcp85_tas = pd.read_csv("data/murg_tas_rcp26_mean.csv").rename(columns={"temp_day": "tas_day_rcp26"})
rcp26_pr   = pd.read_csv("data/murg_pr_rcp85_mean.csv").rename(columns={"precip_mm_day": "pr_day_rcp85"})
rcp85_pr   = pd.read_csv("data/murg_tas_rcp85_mean.csv").rename(columns={"temp_day": "tas_day_rcp85"})

# Sicherstellen, dass "date" als Datetime-Objekt vorliegt
for df in [rcp26_tas, rcp85_tas, rcp26_pr, rcp85_pr]:
    df["date"] = pd.to_datetime(df["date"])

# Zusammenführen über die gemeinsame "date"-Spalte
merged_df = rcp26_tas.merge(rcp85_tas, on="date") \
                      .merge(rcp26_pr, on="date") \
                      .merge(rcp85_pr, on="date")

# Speichern als neue CSV
output_csv = "data/merged_tas_pr_data.csv"
merged_df.to_csv(output_csv, index=False)

print(" Vorschau:")
print(merged_df.head())
```

# Abflussdaten vorbereiten und neu speichern

```python
import pandas as pd

# CSV ab Zeile 7 einlesen mit korrekter Kodierung, da erste 6 Zeilen Projektbeschreibung sind
df = pd.read_csv(
    "/Users/emmamorlock/git/GitHub/projektstudie-murgtal/data/abfluss/Hydrologische-Landespegel-2025-12-22.csv",
    encoding="ISO-8859-1",  
    sep=",",                
    skiprows=6,             
    decimal=",",            
    quotechar='"'          
)

# Anfang anzeigen lassen
print(df.head())
```

```python
# Datum umwandeln
df["Datum / Uhrzeit"] = pd.to_datetime(df["Datum / Uhrzeit"], errors="coerce")

# Nur Datum und Abflusswerte behalten
df = df[["Datum / Uhrzeit", "Wert"]].rename(columns={
    "Datum / Uhrzeit": "date",
    "Wert": "flow_m3_s"
})

print(df.head())
```

```python
# Übersicht anzeigen lassen
df.describe()
```

```python
# Neue Datei speichern als CSV im Ordner data
out_csv = "/Users/emmamorlock/git/GitHub/projektstudie-murgtal/data/murg_abfluss_bereinigt.csv"
df.to_csv(out_csv, index=False, encoding="utf-8")
```
