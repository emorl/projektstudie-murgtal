# Abfluss-, Temperatur- und Niederschlag zusammenführen


```python
#benötigte Pakete laden
import pandas as pd
import os

# working directory setzen
work_dir = r"/Users/emmamorlock/git/GitHub/projektstudie-murgtal"  
os.chdir(work_dir)

# Daten einlesen
era5_tas = pd.read_csv("data/ERA5temperature/daily mean chunks/era5_tas_daily_spatial_mean_1981_2023.csv")#.rename(columns={"precip_mm_day": "pr_day_rcp26"})
era5_prec = pd.read_csv("data/ERA5precipitation/ERA 5 daily sum 1h/era5_prec_daily_spatial_mean_1981_2023.csv")#.rename(columns={"precip_mm_day": "pr_day_rcp85"})
discharge = pd.read_csv("data/murg_abfluss_bereinigt.csv")

# Sicherstellen, dass "date" als Datetime-Objekt vorliegt
for df in [era5_tas, era5_prec, discharge]:
    df["date"] = pd.to_datetime(df["date"])

# Zusammenführen über die gemeinsame "date"-Spalte
merged_df = era5_tas.merge(era5_prec, on="date") \
                      .merge(discharge, on="date") #\
                      #.merge(rcp26_pr, on="date") \
                      #.merge(rcp85_pr, on="date")

# Speichern als neue CSV
output_csv = "data/merged_era5_tas_prec_discharge.csv"
merged_df.to_csv(output_csv, index=False)

# check
print(merged_df.head())
```
